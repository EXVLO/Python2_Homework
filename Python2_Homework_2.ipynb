Write an try except block covering the line that executes the `sum_two_words()` function. Handle at least 2 possible errors that might happen. It should print the returning value when an error is not presented. At last, it should always print `Ve and Due are the cutest` regardless the scenario.

```python
def sum_two_words(w1: str, w2: str) -> str:
    try:
        result = w1 + " " + w2
        print(result)
        return result
    except Exception:
        print(f"An error occurred")
        return ""
    finally:
        print("Ve and Due are the cutest")

w1 = 5 # This can be changed
w2 = 'Rujipas' # This can be changed

sum_two_words(w1, w2)
```

Write function `get_country_name(phone_number)`.

phone_number should be in the format of '+XXOOOOOOOOOO' where `+XX` is the country code and other ten `O` characters is the number itself.
This function should return string values:
- "Spain" if it begins with "+34"
- Your native country if it begins with `your_country_code`
- Optinal: contries form https://en.wikipedia.org/wiki/List_of_country_calling_codes if you are determined enough


Raise:
- LookupError if the number doesn't contain any country code.
- ValueError if the number contains country code, but the amount of characters are incorrect
- TypeError if the telephone number is not a string
- Or other exception that might happen

https://docs.python.org/3/library/exceptions.html  - List of Python build-in exception types

```python
def get_country_name(phone_number: str) -> str:
    if not isinstance(phone_number, str):
        raise TypeError("Phone number must be a string.")

    if not phone_number.startswith('+'):
        raise LookupError("Phone number must start with a '+'")

    if len(phone_number) < 12 or not phone_number[1:].isdigit():
        raise ValueError("Phone number format invalid")

    country_codes = {
        '+34': 'Spain',
        '+995': 'Georgia',
        '+1': 'USA',
        '+49': 'Germany',
        '+81': 'Japan',
        '+39': 'Italy',
    }

    for code, country in country_codes.items():
        if phone_number.startswith(code):
            remaining = len(phone_number) - len(code)
            if remaining != 10:
                raise ValueError("Phone number should contain exactly 10 digits after country code.")
            return country

    raise LookupError("Unknown country code.")

print(get_country_name("+341234567890"))  # Spain
print(get_country_name("+9951234567890"))  # Georgia
print(get_country_name("+11234567890"))   # USA
```

```python
def get_country_name(phone_number: str) -> str:
    assert isinstance(phone_number, str), "The phone number must be a string"
    assert phone_number.startswith('+'), "The phone number must start with a '+'"
    assert len(phone_number) > 1 and phone_number[1:].isdigit(), "The phone number must contain digits after '+'"
    assert len(phone_number) >= 12, "The phone number must include a country code and 10 digits"

    country_codes = {
        '+34': 'Spain',
        '+995': 'Georgia',
    }

    for code, country in country_codes.items():
        if phone_number.startswith(code):
            remaining = len(phone_number) - len(code)
            assert remaining == 10, f"The number part must have exactly 10 digits after '{code}'"
            return country

    raise LookupError("Unknown or unsupported country code")

print(get_country_name("+341234567890"))  # Spain
print(get_country_name("+9951234567890"))  # Georgia
```
