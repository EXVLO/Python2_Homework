1. Write a decorator `n_times` which given a number `n` makes a function to execute `n` times.

```python
def n_times(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator

@n_times(3)
def print_my_name():
    print('Due')

print_my_name()
```

2. Write a decorator `once` which makes a function to execute only once and other times it just returns the result calculated on the first call.

```python
import random

def once(func):
    result = None
    has_run = False

    def wrapper(*args, **kwargs):
        nonlocal result, has_run
        if not has_run:
            result = func(*args, **kwargs)
            has_run = True
        return result

    return wrapper

@once
def random_number():
    return random.randint(1, 10)

print(random_number())
print(random_number())
print(random_number())
```
